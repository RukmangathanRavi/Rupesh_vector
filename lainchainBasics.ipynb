```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-5-nano",
)
```

```python
response  = llm.invoke([
     {"role": "user", "content": "Hello Rupesh"}
])
```

```python
response.content
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()

parser.invoke(response)
```

```python
def upper(val):
    return val.upper()
```

```python
chain = llm | parser | upper

chain.invoke([
     {"role": "user", "content": "Hello Rupesh"}
])
```

```python
from langchain_core.messages import SystemMessage, AIMessage, HumanMessage

prompt = [
    SystemMessage(content="You are a python developer"),
    HumanMessage(content="How to swap 2 list values")
]

response = llm.invoke(prompt)
print(response)
```

```python
def normal(text):
    return text.upper()
```

```python
from langchain_core.runnables import RunnableLambda

converted = RunnableLambda(normal)

converted.invoke("Hello")
```
