```python
from pydantic import BaseModel
from langchain_openai import ChatOpenAI
from langgraph.graph import START, END, StateGraph
from typing import Annotated
from langgraph.graph.message import add_messages
from langgraph.checkpoint.memory import InMemorySaver


class LocalState(BaseModel):
    message : Annotated[list, add_messages]

openAi = ChatOpenAI(model="gpt-4.1")

builder = StateGraph(LocalState)

def node_local(state:LocalState) -> LocalState:
    print(state)
    responseResult = openAi.invoke(state.message)
    # print(responseResult)
    state.message = [{"role":"ai", "content":responseResult.content}]
    return state

builder.add_node("nodelocal", node_local)   
builder.add_edge(START, "nodelocal")
builder.add_edge("nodelocal", END)

result = builder.compile(checkpointer=InMemorySaver())
```

```python

res = result.invoke({"message" : [{"role": "user", "content": "Do you know my name"}]},
              {"configurable": {"thread_id" : "1"}})
```
